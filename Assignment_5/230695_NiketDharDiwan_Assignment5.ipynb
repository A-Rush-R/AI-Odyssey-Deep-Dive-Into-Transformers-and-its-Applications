```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import LSTM, Dense, Embedding
import numpy as np
# Load the dataset
df = pd.read_csv('indian_names.csv')

# Preprocess the data
names = df['Name'].astype(str).str.lower()   # Convert to lowercase

# Split the data into training and testing sets
train_names, test_names = names[:8000], names[8000:]  # Adjust the split ratio as needed

# Create a character index and reverse index
chars = sorted(set(''.join(names)))
char_index = {c: i for i, c in enumerate(chars)}
index_char = {i: c for i, c in enumerate(chars)}
```

```python
# Build the model
model = Sequential()
model.add(Embedding(input_dim=len(chars), output_dim=100, input_length=1))
model.add(LSTM(100))
model.add(Dense(len(chars), activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam')
```

```python
x_train = []
y_train = []
for name in train_names:
    for i in range(len(name) - 1):
        x_train.append(char_index[name[i]])
        y_train.append(char_index[name[i + 1]])

x_train = np.array(x_train)

y_train_categorical = to_categorical(y_train, num_classes=len(chars))
```

```python
# Train the model
model.fit(x_train, y_train_categorical, epochs=100)  # Adjust the number of epochs as needed
```

```python
# Generate new names
def generate_name(start_letter, length):
    input_ = np.array([char_index[start_letter]]) # Convert input to a NumPy array
    name = ''
    for i in range(length):
        predictions = model.predict(input_, verbose=0) # Suppress prediction output
        next_char = index_char[np.argmax(predictions)]
        name += next_char
        input_ = np.array([char_index[next_char]]) # Update input for next prediction
    return name

# Example usage:
for _ in range(10):
    start_char = np.random.choice(chars)
    print(generate_name(start_char, 7))
```
