Averaging filter: It is used in reduction of the detail in image. All coefficients are equal.
<br>
Gaussian filter: A Gaussian Filter is a low pass filter used for reducing noise and blurring regions of an image<br>
Median Filtering: It is also known as nonlinear filtering. It is used to eliminate salt and pepper noise. Here the pixel value is replaced by the median value of the neighboring pixel.<br>
Unsharp Masking: USM is a technique developed to sharpen images, using a blurer or unsharp version of the same image and then, in the common tongue, subtracting it from the original image thus, sharpening the actual image<br>
Prewitt filter: It is uded fordetecting edges in image that are mhorizontal or verticla<br>
Sobel Operator: It is a discrete differentiation gradient-based operator. It computes the gradient approximation of image intensity function for image edge detection <br>
Laplacian filter: It is a second-order derivative filter used in edge detection, in digital image processing. In 1st order derivative filters, we detect the edge along with horizontal and vertical directions separately and then combine both

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread('adu.jpg', cv2.IMREAD_GRAYSCALE)

prewitt_horizontal = cv2.filter2D(image, -1, np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]]))
prewitt_vertical = cv2.filter2D(image, -1, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))

combined_prewitt = cv2.addWeighted(prewitt_horizontal, 0.5, prewitt_vertical, 0.5, 0)
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(combined_prewitt, cmap='gray'), plt.title('Combined Prewitt Filter')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('adu.jpg', cv2.IMREAD_GRAYSCALE)

sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

sobel_combined = cv2.addWeighted(cv2.convertScaleAbs(sobel_x), 0.5, cv2.convertScaleAbs(sobel_y), 0.5, 0)
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(sobel_combined, cmap='gray'), plt.title('Sobel Filter')
plt.show()
```

```python
from tensorflow.keras.datasets import mnist
from matplotlib import pyplot as plt
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
print(f"No. of elements in training dataset: {X_train.shape[0]}")
print(f"No. of elements in testing dataset: {X_test.shape[0]}")
```

```python
from tensorflow.keras.utils import to_categorical
y_train_ohe = to_categorical(y_train)
y_test_ohe = to_categorical(y_test)
```

```python
print("Pixel array of the first image in the training dataset:")
print(X_train[0])
plt.imshow(X_train[0], cmap='gray')
plt.title(f"Label: {y_train[0]}")
plt.show()
```

```python
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
```

```python
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer=Adam(),
              loss=SparseCategoricalCrossentropy(),
              metrics=['accuracy'])
```

```python
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
history = model.fit(X_train_split, y_train_split, epochs=10, batch_size=32, validation_data=(X_val_split, y_val_split))
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.show()
```

```python
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
```

```python
print(classification_report(y_test, y_pred_classes))
```

```python
import random
random_indices = random.sample(range(X_test.shape[0]), 10)
selected_images = X_test[random_indices]
selected_labels = y_test[random_indices]
selected_preds = y_pred_classes[random_indices]
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(selected_images[i].reshape(28, 28), cmap='gray')
    ax.set_title(f"Actual: {selected_labels[i]}\nPred: {selected_preds[i]}")
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
from tensorflow.keras.models import Model
#ngl had to use gpt here
layer_outputs = [layer.output for layer in model.layers]
activation_model = Model(inputs=model.input, outputs=layer_outputs)

first_test_image = X_test[0].reshape(1, 28, 28, 1)
activations = activation_model.predict(first_test_image)

def visualize_layer_output(activations, layer_index, num_filters=32, cols=8):
    activation = activations[layer_index]
    num_filters = min(num_filters, activation.shape[-1])
    rows = num_filters // cols

    plt.figure(figsize=(20, rows * 2.5))
    for i in range(num_filters):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(activation[0, :, :, i], cmap='gray')
        plt.axis('off')
    plt.show()
print("Outputs of the first convolutional layer:")
visualize_layer_output(activations, layer_index=0, num_filters=32)
print("Outputs of the first max pooling layer:")
visualize_layer_output(activations, layer_index=1, num_filters=32)
print("Outputs of the second convolutional layer:")
visualize_layer_output(activations, layer_index=2, num_filters=64)
print("Outputs of the second max pooling layer:")
visualize_layer_output(activations, layer_index=3, num_filters=64)
```
